```python
data = pd.read_csv('dataset/data_cleaned.csv')
data.info()
```

### Encoding `employment_type`

```python
df = data['employment_type']
df = df.str.get_dummies(sep = ',')
df
```

```python
data = pd.concat([data, df], axis=1)
data.drop(['employment_type'], axis = 1, inplace = True)
data.head()
```
